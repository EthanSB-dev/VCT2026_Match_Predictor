```python
import pandas as pd
import glob
import os

data_dir = 'data/raw/'
csv_files = glob.glob(os.path.join(data_dir, '*.csv'))
print(f"Total files: {len(csv_files)}")

# Per-file inspection
for f in sorted(csv_files):
    df = pd.read_csv(f)
    print(f"\n{os.path.basename(f)}: {df.shape[0]} rows, {df.shape[1]} cols")
    print(df.columns.tolist()[:10], '...' if len(df.columns) > 10 else '')
    print(df.dtypes.value_counts())
    print("Missing %:", round(df.isnull().mean().mean() * 100, 2))

# Combine all files
dfs = []
for f in csv_files:
    temp = pd.read_csv(f)
    temp["_source_file"] = os.path.basename(f)
    dfs.append(temp)

combined = pd.concat(dfs, ignore_index=True)
combined.to_csv("combined_output.csv", index=False)

# Global summary
print(f"\n✅ Combined shape: {combined.shape}")
print("\n--- Missing per column ---")
print(combined.isnull().mean().mul(100).round(2).to_string())
print("\n--- Numeric stats ---")
print(combined.describe().round(2).to_string())
```

Project Research Questions
